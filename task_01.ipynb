```python
# Importing Dependancies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Importing the dataset
dataset = pd.read_csv('./dataset/API_SP.POP.TOTL_DS2_en_csv_v2_84031.csv',skiprows=4)
```

```python
# Displaying the dataset
dataset.head()
```

```python
# Describing the dataset
dataset.describe()
```

```python
# Checking for missing values
dataset.isnull().sum()
```

```python
# Dropping the missing values
dataset.dropna(inplace=True)
```

```python
# Selecting countries and years
countries = ['India', 'China', 'United States', 'Indonesia', 'Pakistan']
years = [2000, 2005, 2010, 2015, 2019]
```

```python
# Filtering the dataset
dataset = dataset[dataset['Country Name'].isin(countries)]
dataset = dataset[['Country Name'] + [str(year) for year in years]]
dataset = dataset.set_index('Country Name')
dataset = dataset.T
dataset.index = dataset.index.map(int)

# Plotting the dataset
plt.style.use('ggplot')
dataset.plot(kind='line', figsize=(10, 6))
plt.title('Population of Countries Over Time')
plt.ylabel('Population')
plt.xlabel('Years')
plt.show()

# Creating Bar Graph
dataset.plot(kind='bar', figsize=(10, 6))
plt.title('Population of Countries Over Time')
plt.ylabel('Population')
plt.xlabel('Years')
plt.show()


```
